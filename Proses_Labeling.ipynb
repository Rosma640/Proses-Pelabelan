# **Pelabelan Data**

```python
# Pelabelan data ulasan
def pelabelan(score):
  if score < 4:
    return 'Negatif'
  elif score > 3 :
    return 'Positif'
content['Label'] = content ['score'].apply(pelabelan)
content.head(10)
```
